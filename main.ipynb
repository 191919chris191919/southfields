```python
import torch
import json
import xarray
import pandas
import matplotlib.pyplot
import cartopy.crs
import pathlib
import os
torch.__version__
```

```python
with open ('config.json','r') as tutorial_config:
    tutorial_config = json.load(tutorial_config)
tutorial_config
```

```python
def get_platform_dir(select_platform, config):
    try:
        root_path = pathlib.Path(config['default_dirs'][select_platform]) / ''
    except KeyError:
        root_path = pathlib.Path(os.environ['HOME']) / ''
    return root_path
```

```python
current_platform = tutorial_config['platform']
current_platform
```

```python
root_data_dir = get_platform_dir(current_platform, tutorial_config)

print(root_data_dir.is_dir())
root_data_dir
```
